# 静岡の地価を観察

データは[ふじのくにオープンデータカタログ](http://open-data.pref.shizuoka.jp/)の[平成25年静岡県地価調査結果](http://open-data.pref.shizuoka.jp/muotxqb3a-139/#_139)から

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d = pd.read_csv("shiz.csv")
```

```python
d.head()
```

```python
d.describe()
```

```python
d["価格(円/平方メートル)"]  = d["価格(円/平方メートル)"].str.replace(r'\D+', '').astype('int')
```

```python
d.describe()
```

```python
plt.hist(np.log(d["価格(円/平方メートル)"]))
```

```python
d['所在地']  = d['所在地'].str.replace(r'市.+', '')
d['所在地'] = d['所在地'].str.replace(r'町.+', '')
```

```python
d2 = d.groupby('所在地')
```

```python
d2["価格(円/平方メートル)"].mean().order()
```

```python
d2["価格(円/平方メートル)"].mean().hist()
```
